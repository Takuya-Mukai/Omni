````python
# 1. ライブラリのインポート
import numpy as np
````

````python
# 2. homework関数の定義
# !!WRITE ME!!に処理を記入する（homework関数を提出することに注意）
def homework(a):
    my_result = np.array([])
    for i in a:
        if i % 5 == 0:
            if (i-1) % 2 == 0:
                my_result = np.append(my_result, i)
    return my_result
````

````python
# 3. Numpyの1次元配列の定義
a = np.array([1, 5, 10, 3, 4, 25, 30])
````

````python
# 4. homework関数の出力をprintなどで確認
print(homework(a))
````
